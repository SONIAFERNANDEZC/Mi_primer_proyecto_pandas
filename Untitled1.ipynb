```python
import numpy as np
import seaborn as sns
import matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_context("poster")
sns.set(rc={"figure.figsize": (20.,12.)})
sns.set_style("whitegrid")
```

```python
datafv = pd.read_csv("datafv.csv", encoding="ISO-8859-1")
```

```python
datafv["Sex "] = datafv["Sex "].replace(["lli", ".", ' occupants: T & G Longhi', ' Stephanie', ' occupants 2 men', '  Douglas Richards & George Irwin', '3 fishermen', ' occupants - George Bridge & 2 sons', 'nan', ' occupants - 2 young men', 'Carl Bruster', ' occupants Harry Goodson & Douglas Barnes', 'N', ' occupants: John Wright & a friend'], "Unknown")
```

```python
datafv["Sex "] = datafv["Sex "].replace(["M "], "M")
```
